## QUESTION 8

Scrape the details of Highest selling novels. Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/ You have to find the following details: A) Book name B) Author name C) Volumes sold D) Publisher E) Genre

```python
def uci(url):
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    time.sleep(3)
    link=[x.get_attribute('href') for x in driver.find_elements_by_xpath("//td/span[@class='whitetext']/a")]
    driver.get(link[4])
    time.sleep(3)
    name = [x.text for x in driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td")]
    df = pd.DataFrame({})
    df["Dataset_name"]=name[0::7]
    df['Data_type']=name[1::7]
    df['Task']=name[2::7]
    df['Attribute_type']=name[3::7]
    df['No_of_instances']=name[4::7]
    df['No of attribute']=name[5::7]
    df['Year']=name[6::7]
    df.drop(labels=0,axis =0, inplace =True)
    driver.close()
    return(df)

uci('https://archive.ics.uci.edu/')
```

## QUESTION 9

Scrape the details most watched tv series of all time from imdb.com. Url = https://www.imdb.com/list/ls095964455/ You have to find the following details: A) Name B) Year span C) Genre D) Run time E) Ratings F) Votes

```python
def series(url):
    driver=webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    name = [x.text for x in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")]
    year_span= [x.text for x in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")]
    genre = [x.text for x in driver.find_elements_by_xpath("//span[@class='genre']")]
    run_time = [x.text for x in driver.find_elements_by_xpath("//span[@class='runtime']")]
    ratings = [x.text for x in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")]
    votes = [x.text for x in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[@name='nv']")]
    driver.close()
    df = pd.DataFrame({})
    df['name'] =name 
    df['year_span']=year_span 
    df['genre'] =genre 
    df['run_time'] = run_time
    df['ratings'] =ratings 
    df['votes'] =votes
    return(df)



series('https://www.imdb.com/list/ls095964455/')
```
